### 1. Load Data

```python
from pybug.io import auto_import
from pybug.landmark.labels import \
    (labeller, 
     ibug_68_points,
     ibug_68_contour,
     ibug_68_trimesh)
from pybug.shape import PointCloud


# load the training images of the LFPW database as landmarked images using the autoimporter
images = auto_import('/vol/atlas/databases/lfpw/testset/' + '*.png', max_images=10)

# label the landmarks using the ibug's "standard" 68 points mark-up
labeller(images, 'PTS', ibug_68_points)
labeller(images, 'PTS', ibug_68_contour)
labeller(images, 'PTS', ibug_68_trimesh)

# Extract shape data. -1 because the annotation are 1 based
points = [img.landmarks['PTS'].all_landmarks.points - 1  for img in images]
shapes = [PointCloud(p) for p in points]
```

```python
images[0].get_landmark_set("ibug_68_points").view_new()
gcf().set_size_inches((10,10))
```

### 2. Load AAM

```python
import pickle

# load a previously buid AAM
aam = pickle.load(open('/vol/atlas/aams/lfpw_pwa', "rb"))

reference_frame = aam["template"]
appearance_model = aam["appearance_model"]
shape_model = aam["shape_model"]
```

### 3. Build Fitters

```python
from pybug.transform.piecewiseaffine import PiecewiseAffineTransform
from pybug.transform.affine import SimilarityTransform
from pybug.transform.statisticallydriven import StatisticallyDrivenTransform

source = reference_frame.landmarks['PTS'].all_landmarks
template = appearance_model.mean

trilist = reference_frame.get_landmark_set('ibug_68_trimesh').landmark_dict['tri'].trilist
def pwa_constructor(src_landmarks, tgt_landmarks):
    return PiecewiseAffineTransform(src_landmarks, tgt_landmarks, trilist)

dummy_similarity_transform = SimilarityTransform.from_vector(np.array([0, 0, 0, 0]))
global_transform = dummy_similarity_transform

sta_driven_transform = StatisticallyDrivenTransform(shape_model, pwa_constructor,
                                                    source=source, 
                                                    global_transform=global_transform)
```

```python
from pybug.lucaskanade.residual import LSIntensity
from pybug.lucaskanade import \
    (ProjectOutInverseCompositional,
     SimultaneousInverseCompositional,
     AdaptiveInverseCompositional,
     AlternatingInverseCompositional,
     ProjectOutForwardCompositional,
     ProbabilisticForwardCompositional,
     ToSubspaceForwardCompositional,
     WithinSubspaceForwardCompositional)

# standard Least Squares Residual
residual = LSIntensity()


# Project Out Inverse Compositional
project_out = ProjectOutInverseCompositional(appearance_model, 
                                             residual, 
                                             sta_driven_transform)
# Project Out Inverse Compositional
simultaneous = SimultaneousInverseCompositional(appearance_model, 
                                                residual, 
                                                sta_driven_transform)
# Adaptive Inverse Compositional
adaptive = AdaptiveInverseCompositional(appearance_model, 
                                        residual, 
                                        sta_driven_transform)
# Alternating Inverse Compositional
alternating = AlternatingInverseCompositional(appearance_model, 
                                              residual, 
                                              sta_driven_transform)


# Project Out Forward Compositional
cone = ProjectOutForwardCompositional(appearance_model, 
                                      residual, 
                                      sta_driven_transform)

# Project Out Forward Compositional (GD)
#code = ProjectOutForwardCompositional(appearance_model, 
#                                      residual, 
#                                      sta_driven_transform,
#                                      optimisation=('GD',0))


# Probabilistic Forward Compositional
probabilistic = ProbabilisticForwardCompositional(appearance_model, 
                                                residual, 
                                                sta_driven_transform)
# To Subspace Forward Compositional
to_subspace = AlternatingInverseCompositional(appearance_model, 
                                              residual, 
                                              sta_driven_transform)
# Within Subspace Forward Compositional
within_subspace = AlternatingInverseCompositional(appearance_model, 
                                                  residual, 
                                                  sta_driven_transform)
```

### 4. Compute Initializations

```python
from pybug.transform.affine import SimilarityTransform
import numpy as np

noise_scale = 0.1

# estimate the true similarity transform
ori_sim_trans = [SimilarityTransform.estimate(shape_model.mean.points, s.points) 
                for s in shapes]

ini_sta_params = []
ini_sta_trans = []
for t in ori_sim_trans:
    ini_sim_param = t.as_vector()
    # kill rotation
    ini_sim_param[1] = 0
    for i in range(10):
        # add random gaussian noise to the ground truth parameters
        ini_sim_param_ = ini_sim_param + noise_scale * np.random.randn(4)
        ini_sta_param_ = np.concatenate([ini_sim_param_, 
                                         np.zeros(shape_model.n_components)])
        ini_sta_params.append(ini_sta_param_)
        ini_sta_trans.append(sta_driven_transform.from_vector(ini_sta_param_))
```

```python
from pybug.warp import scipy_warp

i = 5
j = i / 10

initial_warped_image = scipy_warp(images[j], template, ini_sta_trans[i])

images[j].add_landmark_set("initial", {"initial": ini_sta_trans[i].target})
images[j].get_landmark_set("initial").view_new()
gcf().set_size_inches((10,10))

initial_warped_image.add_landmark_set('reference', {'reference': source})
initial_warped_image.get_landmark_set('reference').with_label('reference').view_new()
gcf().set_size_inches((6,6))
```

```python
gs_images = [img.as_greyscale() for img in images]
```
