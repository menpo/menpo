```python
import numpy as np
import csv
```

```python
filename = '/data/db/lfpw/train/image_0001.pts'
ext = str.rsplit(filename, '.', 1)[-1]
f = open(filename, 'r')
strdata = []
```

```python
if ext == 'pts':
    for line in f:
        if line == '{\n':
            break
    for line in csv.reader(f, delimiter = ' '):
        if line != '}':
            strdata.append(line)
```

```python
data = np.array(strdata, dtype=np.float)
```
